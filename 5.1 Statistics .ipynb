# stratified sampling: take random samples from each group (ie position)
- strafied group is called a stratum (pl strata)

```python
wnba['pts per game'] = wnba['PTS'] / wnba['Games Played']
wnba.head()
```

```python
g = wnba[wnba['Pos'] == 'G']
gf = wnba[wnba['Pos'] == 'G/F']
fc = wnba[wnba['Pos'] == 'F/C']
f = wnba[wnba['Pos'] == 'F']
c = wnba[wnba['Pos'] == 'C']
```

```python
dictmeans = {}
for stratum, pos in [(g,'G'),(gf,'G/F'),(fc,'F/C'),(f,'F'),(c,'C')]:
    sample = stratum.sample(10, random_state=0)
    dictmeans[pos] = sample['pts per game'].mean()
dictmeans
```

```python
position_most_points = max(dictmeans, key=dictmeans.get)
position_most_points
```

```python
wnba['Games Played'].value_counts(bins = 3, normalize = True) * 100
```

```python
strata1 = wnba[wnba['Games Played'] <= 12]
strata2 = wnba[(wnba['Games Played'] <= 22) & (wnba['Games Played'] > 12)]
strata3 = wnba[(wnba['Games Played'] > 22)]
```

```python
import pandas as pd
```

```python
sample_means = []
for i in range(0,100):
    sample1 = strata1.sample(1, random_state=i)
    sample2 = strata2.sample(2, random_state=i)
    sample3 = strata3.sample(7, random_state=i)
    samples_concat = pd.concat([sample1, sample2, sample3])
    sample_means.append(samples_concat['PTS'].mean())
```

```python
#this plot sucks, we will use minutes played instead for a better estimate of mean total points
plt.scatter(x=range(1,101), y=sample_means)
plt.axhline(wnba['PTS'].mean())
```

```python
def ordinalize_points(pts):
    if pts <= 20:
        return 'very few points'
    elif pts <= 80:
        return 'few points'
    elif pts <= 150:
        return 'many, but below average'
    elif pts <= 300:
        return 'average number of points'
    elif pts <= 450:
        return 'more than average'
    else:
        return 'much more than average'
```

```python
PTS_ordinal_scale = wnba['PTS'].apply(ordinalize_points)
PTS_ordinal_scale.head()
```

```python
wnba['PTS_ordinal_scale'] = PTS_ordinal_scale
freq = wnba['PTS_ordinal_scale'].value_counts()
freq
```

```python
import numpy as np
```

```python
def Exp_ordinalize(yrs):
    if yrs == 'R':
        return np.nan
    yrs = int(yrs)
    if yrs <= 1:
        return 'Rookie'
    elif yrs <= 3:
        return 'Little experience'
    elif yrs <= 5:
        return 'Experienced'
    elif yrs <= 10:
        return 'Very experienced'
    elif yrs > 10:
        return 'Veteran'

```

```python
wnba['Exp_ordinal'] = wnba['Experience'].apply(Exp_ordinalize)
wnba['Exp_ordinal']
```
